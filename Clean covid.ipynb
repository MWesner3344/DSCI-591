Seems as though, some locations are missing dates within the dataset

```python
#Get all unique locations
location = new['location'].unique()
#Only examine locations with all 278 dates
new_loc = []
delete = []

for x in location:
    if (len(new[new['location'] == x]) < 278):
        delete.append(x)
    else:
        new_loc.append(x)
```

```python
len(new_loc)
```

```python
#Contains all locations that does not have all 278 dates
delete
```

```python
#Contains rows of all 278 dates 
new_loc
```

```python
#Remove those locations within the delete list
for i in delete:
    new = new.drop(new[new['location'] == i].index)
```

```python
#Examine the remaining locations
new['location'].unique()
```

```python
#Retrieve all columns about covid data
col1 = new.columns[4:33]
col1
```

```python
#Retrieve all columns that is solely focused on coronavirus numbers
col2 = new.columns[12:18]
print(col2)
new.loc[new['date'] == '2019-12-31',col2] = 0.0
```

```python
#Front fill numbers within each location
for x in new_loc:
    new.loc[new.location == x, col1] = new[new['location'] == x][col1].ffill(axis=0)
```

```python
new.isnull().sum()
```

```python
new
```

```python
#Back fill if you cant get rid of of missing values
for x in new_loc:
    new.loc[new.location == x, col1] = new[new['location'] == x][col1].bfill(axis=0)
```

```python
new.isnull().sum()
```

Observations:
- The front fill helped to fill in numbers specifically about the coronavirus. But did not within the locations' statistical data about population, age, diabetes, life expectancy, etc. Most likely this dataset does not have any data on these numbers and if needed, we would have to find another dataset to understand these numbers. These numbers cannot be absentmindedly filled with zeros. It would not make sense to have 0 smokers, etc.
